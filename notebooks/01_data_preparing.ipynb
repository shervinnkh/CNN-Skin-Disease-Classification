```python
from ShervinKhLib.ai_tools import train_test_valid_split
from ShervinKhLib import save_var
import numpy as np
import cv2 
import os
```

```python
img = cv2.imread('../data/dataset/train/acne/ac100_JPG.rf.a41d4a786f2197b20cfa6372cc1625a5.jpg')
```

```python
categories = os.listdir('../data/dataset/train/')
```

```python
x=[]
y=[]

for category in categories:
    files = os.listdir('../data/dataset/train/' + category + '/')
    for i,file in enumerate( files):
        img = cv2.imread('../data/dataset/train/' + category + '/' + file)
        x.append(img)
        y.append(category)

        print((i/len(files))*100 , "%")
```

```python
x = np.array(x)
y = np.array(y)
```

```python
x_train , x_test , x_valid , y_train , y_test , y_valid = train_test_valid_split(x,y,train_size=0.7 , test_size=0.15)
```

```python
save_var("../data/vars/x_train",x_train)
save_var("../data/vars/x_test",x_test)
save_var("../data/vars/x_valid",x_valid)
save_var("../data/vars/y_train",y_train)
save_var("../data/vars/y_test",y_test)
save_var("../data/vars/y_valid",y_valid)
```
